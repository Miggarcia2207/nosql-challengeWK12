2. Find the BusinessTypeID for "Restaurant/Cafe/Canteen" and return only the `BusinessTypeID` and `BusinessType` fields.

```python
# Find the BusinessTypeID for "Restaurant/Cafe/Canteen" and return only the BusinessTypeID and BusinessType fields
# Define the query filter
query = {
    'BusinessType': 'Restaurant/Cafe/Canteen'
}

# Define the projection
projection = {
    'BusinessTypeID': 1,
    'BusinessType': 1,
    '_id': 0
}

# Find documents matching the query and apply the projection
results = collection.find(query, projection)

# Iterate over the results and print each document
for result in results:
    print(result)
```

3. Update the new restaurant with the `BusinessTypeID` you found.

```python
# Update the new restaurant with the correct BusinessTypeID
# Find the BusinessTypeID for "Restaurant/Cafe/Canteen"
business_type_query = {
    'BusinessType': 'Restaurant/Cafe/Canteen'
}
business_type_projection = {
    'BusinessTypeID': 1,
    '_id': 0
}
business_type_result = collection.find_one(business_type_query, business_type_projection)

# Check if the business type exists and retrieve the BusinessTypeID
if business_type_result:
    business_type_id = business_type_result['BusinessTypeID']

    # Update the new restaurant document
    update_query = {
        'name': 'Penang Flavours'
    }
    update_values = {
        '$set': {
            'BusinessTypeID': business_type_id
        }
    }
    update_result = collection.update_one(update_query, update_values)

    # Check if the update was successful
    if update_result.modified_count > 0:
        print("The new restaurant has been updated with the correct BusinessTypeID.")
    else:
        print("Failed to update the new restaurant.")
else:
    print("The BusinessTypeID for 'Restaurant/Cafe/Canteen' was not found.")
```

```python
# Confirm that the new restaurant was updated
# Find the updated restaurant document
updated_restaurant = collection.find_one({'name': 'Penang Flavours'})

# Check if the updated document exists
if updated_restaurant:
    print("The new restaurant has been updated successfully.")
    print("Updated Restaurant Details:")
    print(updated_restaurant)
else:
    print("The new restaurant was not found in the collection.")
```

5. Some of the number values are stored as strings, when they should be stored as numbers.

Use `update_many` to convert `latitude` and `longitude` to decimal numbers.

```python
# Change the data type from String to Decimal for longitude and latitude

# Build the update query
update_query = {
    '$or': [
        {'longitude': {'$type': 'string'}},
        {'latitude': {'$type': 'string'}}
    ]
}

# Build the update values
update_values = {
    '$set': {
        'longitude': {
            '$toDouble': '$longitude'
        },
        'latitude': {
            '$toDouble': '$latitude'
        }
    }
}

# Update the documents
update_result = collection.update_many(update_query, update_values)

# Print the number of updated documents
print("Number of updated documents:", update_result.modified_count)
```

Use `update_many` to convert `RatingValue` to integer numbers.

```python
# Set non 1-5 Rating Values to Null
non_ratings = ["AwaitingInspection", "Awaiting Inspection", "AwaitingPublication", "Pass", "Exempt"]
establishments.update_many({"RatingValue": {"$in": non_ratings}}, [ {'$set':{ "RatingValue" : None}} ])
```

```python
# Change the data type from String to Integer for RatingValue
# Build the update query
update_query = {
    'RatingValue': {'$type': 'string'}
}

# Build the update values
update_values = {
    '$set': {
        'RatingValue': {
            '$toInt': '$RatingValue'
        }
    }
}

# Update the documents
update_result = establishments.update_many(update_query, update_values)

# Print the number of updated documents
print("Number of updated documents:", update_result.modified_count)
```

```python
# Check that the coordinates and rating value are now numbers
# Find a document that includes the fields
document = establishments.find_one({'longitude': {'$exists': True}, 'latitude': {'$exists': True}, 'RatingValue': {'$exists': True}})

# Check if the document exists
if document:
    print("The coordinates and RatingValue are now numbers.")
    print("Longitude:", document['longitude'])
    print("Latitude:", document['latitude'])
    print("RatingValue:", document['RatingValue'])
else:
    print("The fields were not updated correctly.")
```
